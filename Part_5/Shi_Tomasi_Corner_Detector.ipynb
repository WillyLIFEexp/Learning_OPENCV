# Shi-Tomasi Corner Detector
> Shi-Tomasi Corner Detector比Harris corner來得強也簡單使用
> 同樣也是可以幫忙定義corner的detection
> 
> 以下是Opencv Doucument 所述的parameter
> ![title](https://snag.gy/56igJT.jpg)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../../Opencv/Computer-Vision-with-Python/DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
real_chess = cv2.imread('../../Opencv/Computer-Vision-with-Python/DATA/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess,cmap='gray')
```

```python
# 在cv2裡面, goodFeaturesToTrack就是Shi Tomasi Corner detection
corners = cv2.goodFeaturesToTrack(gray_flat_chess,5,0.01,10)
# 雖說goodFeaturestotack比Harris 來得方便使用, 但他並不會幫我們將
# 圈圈自動化上去, 所以我們需要自己將圈圈畫上去
corners = np.int0(corners)

print(corners)
print(corners[0].shape)
print(corners[0].ravel().shape)

for i in corners:
    # 將資料轉成 1d array
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,255,10)

plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,80,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),3,255,-1)

plt.imshow(real_chess)
```
