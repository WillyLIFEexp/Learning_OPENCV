# Drawing Basic Shapes

我們要透過cv2在照片上面畫圖

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(121)
    ax.imshow(img,cmap)
```

```python
# 我們先建立黑色的背景, 方便我們等等畫圖
background = np.zeros((512,512,3),np.int16)
display(background)
```

```python
# 正方形
# cv2.rectangle(img,pt1,pt2,color,thickness)
rect = cv2.rectangle(background,(50,50),(200,200),(255,0,0),3)
display(rect)
```

```python
# 圓形
# cv2.circlr(img,pt,radius,color,thickness)
# 如果想要填滿圖案的話將把thickness改成負值即可
cir = cv2.circle(background,(300,300),80,(0,255,0),3)
display(cir)
```

```python
# 線條
#cv2.line(img,pt1,pt2,color,thickness)
l = cv2.line(background,(0,0),(500,500),(255,255,255),5)
display(l)
```

```python
# 文字
# 須先定義好文字的格式
font = cv2.FONT_HERSHEY_COMPLEX
# cv2.putText(img,text,origin,font_style,size,color,thickness,)
# The text origin will start from the bottom left of the text
txt = cv2.putText(background,"Hello",(0,500),font,5,(0,0,255),2,cv2.LINE_4)
display(txt)
```

```python
# Polygon
# 要透過opencv畫出多邊形,我們需要先定義出我們要的點存在 np array裡面
# 也要注意np array 需要將型態轉乘 int32
```

```python
# 我們先建立黑色的背景, 方便我們等等畫圖
background = np.zeros((512,512,3),np.int16)
display(background)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)
```

```python
# 如需要話polygon, 我們需要將np array的形狀透過reshape 改成如下的型態
pts = vertices.reshape((-1,1,2))
pts.shape
```

```python
# cv2.polylines(img,[pts],isClosed,color,thickness)
# isClosed 是看我們要不要把polygon關起來
poly = cv2.polylines(background,[pts],False,(255,0,0),5)
```

```python
display(poly)
```

```python
vertices2 = np.array([[150,350],[250,250],[450,350],[250,450]],np.int32)
pts2 = vertices2.reshape((-1,1,2))
poly = cv2.polylines(poly,[pts2],True,(255,255,255),5)
```

```python
display(poly)
```
