# Color Mapping

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../Test_Photo/cat.jpg")
```

#### 透過 cv2 將照片讀取進來, 但是如果直接使用 plt.imshow 時會發現照片顏色怪怪的

```python
plt.imshow(img)
```

#### 這邊主要的原因是因為當 cv2 將照片讀取進來時, 顏色的順序會從RGB換成BGR. 但對於 plt.imshow來說在讀取時還是用RGB的方式讀取. 所以導致顏色會出錯.
#### 所以需要透過 cv2的轉色功能將照片的順序調整回來再顯示就不會有問題了

```python
img_c = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_c)
```

#### cv2的轉色功能除了可以提供將BGR轉乘RGB以外, 也有提供 HSL (hue, saturation, lightness) 與 HSV (hue, saturation, value)轉換的功能
![title](https://snag.gy/JR8hxs.jpg)
#### 不管是HSL或者HSV都比RGB所提供出來的資訊較多, 像是將顏色的資訊以及顏色的強度區分.

```python
img_c = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img_c)
```

```python
img_c = cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
plt.imshow(img_c)
```
